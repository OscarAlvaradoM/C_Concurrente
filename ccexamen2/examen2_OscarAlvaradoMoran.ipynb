```python
# Implementando con Lock
class Mutex:
    def __init__(self):
        self.esta_bloqueado = False
        self.lock = Lock()
    
    def bloquear(self):
        self.lock.acquire()
        self.esta_bloqueado = True
    
    def desbloquear(self):
        self.lock.release()
        self.esta_bloqueado = False
```
