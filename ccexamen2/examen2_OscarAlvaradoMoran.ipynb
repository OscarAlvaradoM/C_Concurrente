# Programa 2

Implementa un programa secuencial que calcule el producto de dos grandes matrices (de manera convencional, es decir, renglón por columna). Después modifícalo para que esta tarea se realice entre cuatro *threads*, cada uno ocupado de un subconjunto de la matriz resultado. Mide el tiempo que emplea cada una de las versiones.

```python
# Secuencial
def multiplicacion(A, B):
    C = np.zeros((A.shape[0], B.shape[1]))
    for i in range(A.shape[0]):
        for j in range(B.shape[1]):
            for k in range(B.shape[0]):
                C[i,j] += A[i,k] * B[k,j]
    return C


# Con Hilos
def cuadrante1(A, B):
    global C, n, m, l, y1, x1
    for i in range(y1):
        for j in range(l):
            for k in range(x1):
                C[i,j] += A[i,k] * B[k,j]

def cuadrante2(A, B):
    global C, n, m, l, y1, x1
    for i in range(y1):
        for j in range(l):
            for k in range(x1, n):
                C[i,j] += A[i,k] * B[k,j]

def cuadrante3(A, B):
    global C, n, m, l, y1, x1
    for i in range(y1, m):
        for j in range(l):
            for k in range(x1):
                C[i,j] += A[i,k] * B[k,j]

def cuadrante4(A, B):
    global C, n, m, l, y1, x1
    for i in range(y1, m):
        for j in range(l):
            for k in range(x1, n):
                C[i,j] += A[i,k] * B[k,j]
            
m = 400
n = 400
l = 400

A = np.random.randint(0,100, (m,n))
B = np.random.randint(0,100, (n,l))
C = np.zeros((m,l))

t_0 = time()
print(multiplicacion(A, B))
print("Tiempo para el caso secuencial: ", time() - t_0)

t_0 = time()
x1 = int(np.floor(n/2))
y1 = int(np.floor(m/2))

thread1 = Thread(target = cuadrante1, args = (A, B))
thread2 = Thread(target = cuadrante2, args = (A, B))
thread3 = Thread(target = cuadrante3, args = (A, B))
thread4 = Thread(target = cuadrante4, args = (A, B))

thread1.start()
thread2.start()
thread3.start()
thread4.start()

thread1.join()
thread2.join()
thread3.join()
thread4.join()
print(C)
print("Tiempo para el caso con hilos: ", time() - t_0)
```
